```
SOURCE = "/content/drive/My Drive/workNT/h5 models" #FOLDER WHERE .h5 files are located in drive
DESTINATION = "/content/drive/My Drive/workNT/tflite models" #where to save the tflite files

files = os.listdir(SOURCE)
print(files)

for file in files:
  name = file[0:len(file) - 3]
  print(name)
  if(file.endswith('.h5')):
    model = tf.keras.models.load_model(SOURCE+'/'+file)
    converter = tf.lite.TFLiteConverter.from_keras_model(model)
    tflite_model = converter.convert()
    open(DESTINATION+"/"+name+".tflite", "wb").write(tflite_model)

print("Finished converting models")
```
